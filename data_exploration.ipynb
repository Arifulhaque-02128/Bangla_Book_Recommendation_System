```python
import pandas as pd
```

```python
#load dataset
books = pd.read_csv('Dataset/Dataset1/refined_bangla_dataset/bangla_train.csv')
```

```python
books
```

```python
import re
```

```python
import emoji
import string

def remove_emojis(text):
    return emoji.replace_emoji(text, "")

def preprocessing(text):
    # Remove emojis
    text = remove_emojis(text)

    # Remove extra whitespace
    text = " ".join(text.split()).strip()

    # Remove punctuations (keep Bangla characters, numbers, and essential symbols)
    text = re.sub(f"[{string.punctuation}]", "", text)

    # Remove non-Bangla characters (keeping Bangla letters, digits, and spaces)
    text = re.sub(r"[^\u0980-\u09FF\s\d]", "", text)

    return text
```

```python
clean_book_names = []
for book in books["Book_Name"] :
  clean_book_names.append(preprocessing(book))


clean_categories = []
for cat in books["Category"] :
  clean_categories.append(preprocessing(cat))
```

```python
clean_book_names
```

```python
clean_categories
```

```python
books["Category"] = clean_categories
books['Book_Name'] = clean_book_names
```

```python
books
```

```python
books['Category'].value_counts()
```

```python
books['Book_Name'].value_counts()
```

```python
clean_writer_names = []
for writer in books["Writer_Name"] :
  clean_writer_names.append(preprocessing(writer))


clean_reviews = []
for review in books["Review"] :
  clean_reviews.append(preprocessing(review))
```

```python
clean_writer_names
```

```python
clean_reviews
```

```python
books["Review"] = clean_reviews
books["Writer_Name"] = clean_writer_names
```

```python
books
```

```python
books.loc[:, 'words_in_review'] = books['Review'].str.split().str.len()
```

```python
books['words_in_review'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns_of_interest = ['Rating', 'label', 'words_in_review']

correlation_matrix = books[columns_of_interest].corr(method='spearman')

sns.set_theme(style="white")
plt.figure(figsize = (8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={"label": "Spearman correlation"})

heatmap.set_title('Correlation Heatmap')
plt.show()
```
